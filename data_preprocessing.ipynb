```python
categorical_columns=df.select_dtypes(include=['object']).columns
categorical_columns
```

```python
#label encoding
from sklearn.preprocessing import LabelEncoder

label_encoder=LabelEncoder()#initializing
df['Species']=label_encoder.fit_transform(df['Species'])#fitting the column
df
```

```python
'''Label encoding assigns each category an integer value.
In this case,
Iris-setosa = 0
Iris-versicolor = 1
Iris-virginica = 2
It as tool Nan into consideration as 3'''

```

```python
df['Species']=df['Species'].replace(3,np.nan)
df.head()
```

```python
#mode imputation for filling missing values for categorical columns
#mode=most frequent value
mode=df['Species'].mode()[0] #[0] is used to access the first (and typically only) mode.
df['Species'].fillna(mode,inplace=True)
df
```

```python
#mean/median values for numerical columns
mean=df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']].mean()
print(mean)
```

```python
df=df.fillna(mean)
df
```
