```python
#replace missing values with mean/median/mode/etc
import numpy as np
df1 = pd.DataFrame({'A': [1, 2, np.nan, 4],
                   'B': [5, np.nan, 7, 8],
                   'C': [np.nan, 10, 11, 12]})
df1
```

```python
r=df1.fillna(df1.mean(),inplace=False)
r
```

```python
#drop columns with missing values
r=df1.dropna(axis=1,inplace=True)
r
```

```python
#rename columns
df.rename(columns={'children':'kids'},inplace=True)
df
```

```python
#combine datasets
df1 = pd.DataFrame({'key': ['A', 'B', 'C', 'D'],
                    'value': [1, 2, 3, 4]})

df2 = pd.DataFrame({'key': ['B', 'D', 'E', 'F'],
                    'value': [5, 6, 7, 8]})

merged_df = pd.merge(df1, df2, on='key')
merged_df
```
